```python
h = open('county_pop.json',) 
# returns JSON object as  
# a dictionary 
data2 = json.load(h)
```

```python
pop_df = json_normalize(data2)
pop_df = pop_df.rename(columns = {"County":"countycombo"})
pop_df.head()
```

```python
pop_df[['County','state']] = pop_df.countycombo.str.split(",",expand=True,)
pop_df
```

```python
counties_new_list = []
test_list = pop_df['County'].to_list()
for item in test_list:
    split = item.rsplit(' ', 1)
    split = split[0]
    counties_new_list.append(split)
counties_new_list
```

```python
#pop_df['County'] = pop_df['County'].str[:-7]
```

```python
pop_df['County'] = counties_new_list
```

```python
pop_df['state'] = pop_df['state'].str[1:]
```

```python
pop_list = pop_df['County'].unique()
```

```python
nyt_list = county_with_ac['County'].unique()
```

```python
pop_to_merge = pop_df[['County','state','Population']]
pop_to_merge = pop_to_merge.groupby(['County', 'state'], as_index = False).sum()
pop_to_merge.head()
```

```python
county_to_merge = pd.merge(county_with_ac, pop_to_merge, on =['County','state'], how='left')
county_to_merge
```

```python
county_to_merge = county_to_merge.rename(columns = {"County": "properties.name", "Acronym":'properties.stusab'})
county_to_merge['death_pop_rate'] = county_to_merge['deaths']/county_to_merge['Population']
county_to_merge['death_case_rate'] = county_to_merge['deaths']/county_to_merge['cases']
county_to_merge['case_pop_rate'] = county_to_merge['cases']/county_to_merge['Population']
```

```python
county_to_merge
```

```python
county_to_merge = county_to_merge.rename(columns = {'properties.name':'County'})
county_to_merge = county_to_merge[county_to_merge.County != 'Unknown']
recent_county = county_to_merge.loc[county_to_merge['date']==county_to_merge['date'].max()]
#nyc
recent_county.at[188943, 'fips'] = 36061.0
recent_county.at[188943, 'Population'] = 8700000
# #kansas city
recent_county.at[323459, 'fips'] = 38000.0
recent_county.at[323459, 'Population'] = 505198
# #joplin co
recent_county.at[323460, 'fips'] = 37592.0
recent_county.at[323460, 'Population'] = 180046
recent_county = recent_county.astype({'fips':'int64'})

recent_county['death_pop_percent'] = recent_county['deaths']/recent_county['Population']*100
recent_county['death_case_percent'] = recent_county['deaths']/recent_county['cases']*100
recent_county['case_pop_percent'] = recent_county['cases']/recent_county['Population']*100

recent_county['death_pop_per_10000'] = recent_county['deaths']/recent_county['Population']*10000
recent_county['death_case_per_10000'] = recent_county['deaths']/recent_county['cases']*10000
recent_county['case_pop_per_10000'] = recent_county['cases']/recent_county['Population']*10000

recent = recent_county.loc[recent_county['fips'].isnull()]
recent_county
```

```python
nyc = recent_county.loc[recent_county['County'] == 'New York City']
nyc
```

```python
final_county = recent_county[['County','state','fips','cases','deaths','death_pop_percent','death_case_percent','case_pop_percent','death_pop_per_10000','death_case_per_10000','case_pop_per_10000']]
final_county
```
