# Country Wide Gas Data

```python
info = requests.get(url).json()
```

```python
info
```

```python
prices = info['series'][0]['data']
prices
```

```python
dates = []
costs = []
for price in prices:
    date = price[0]
    dates.append(date)
    cost = price[1]
    costs.append(cost)
```

```python
dates
```

```python
for date in dates:
    date = int(date)
dates
```

```python
good_dates = []
for s in dates:
    s = datetime.strptime(s, '%Y%m')
    good_dates.append(s)
final_dates = []
final_costs = []
i = 0
while i < 12:
        final_dates.append(good_dates[i])
        final_costs.append(costs[i])
        i = i+1
```

```python
plt.plot(final_dates, final_costs)
plt.xticks(rotation=45)
```

```python
import time
for state in states:
    headers = {
    'content-type': "application/json",
    'authorization': gasAPI
    }

    conn.request("GET", f"/gasPrice/stateUsaPrice?state={state}", headers=headers)
    time.sleep(1)
    res = conn.getresponse()
    data = res.read()
    data = data.decode("utf-8")
    state_gas_data.append(data)

```

```python
states = []
gas_prices = []
for data in state_gas_data:
    hope = json.loads(data)
    state_name = hope['result']['state']['name']
    gas_price = hope['result']['state']['gasoline']
    states.append(state_name)
    gas_prices.append(gas_price)
```

```python
print(states)
print(gas_prices)
```

```python
states = states[1:]
gas_prices = gas_prices[1:]
print(states)
print(gas_prices)
```

```python
dollars = []
for price in gas_prices:
    price = float(price)
    dollars.append(price)
```

```python
plt.bar(states, dollars)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize']=(40,10)
plt.savefig('gas.png')
plt.show()
```
