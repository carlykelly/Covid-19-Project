```python
import json
import pandas as pd
```

## COVID Population Data

```python
# Opening JSON file 
f = open('state_pop.json',) 
# returns JSON object as  
# a dictionary 
data = json.load(f) 
```

```python
#data
```

```python
state_population=[]
# Iterating through the json 
# list 
for i in data: 
    #print(i)
    state_population.append(i)
```

## COVID STATE DATA

```python
# Opening JSON file 
f = open('nyt_covid_states.json',)
# returns JSON object as  
# a dictionary 
data = json.load(f) 
```

```python
state_COVID=[]
# Iterating through the json 
# list 
for i in data: 
    #print(i)
    state_COVID.append(i)
```

```python
# Closing file 
f.close() 
```

```python
state_pop_df=pd.DataFrame(state_population)
state_pop_df.head(5)
```
